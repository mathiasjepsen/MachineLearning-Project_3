## Part 1 - Data Exploration & Preparation

```python
%pylab inline
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import seaborn as sns


voice_csv = pd.read_csv("voice.csv")
x = voice_csv.iloc[:,0:-1]
y = LabelEncoder().fit_transform(voice_csv.iloc[:,-1])


plt.figure(figsize=(30, 24))
corr = x.corr()
sns.heatmap(x.corr(), 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
            cmap=sns.color_palette("Blues"),
            annot=True)
```

If we take a closer look at the correlation matrix plot shown above, we see that the six columns sd, IQR, skew, kurt, sp. ent and sfm are considerably more anti-correlated than the rest of the columns (since their values are negative). Due to these columns not contributing any meaningful values to the overall model, we decided to attempt leaving them out in hopes it would improve our accuracy score with the different classifiers.

```python
from sklearn.neighbors import KNeighborsClassifier


x = x.drop(columns=["sd", "IQR", "skew", "kurt", "sp.ent", "sfm"]);
```

## Part 2: Model construction

### Part 2.1: k-Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.model_selection import cross_val_score

#we make a pipeline containing the KNeighborsClassifier and a starndardScaler.
pipeline1 = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=5))
scores = cross_val_score(pipeline1,x, y, cv=10)
sum(scores/len(scores))
```

The acurracy score of the KNeighborsClassifier is 0.9527087811479978. This is only 0.03% better than our original score(The score from project-2). This tell us that the features we removed niether have a positiv or negativ effect on the accuracy of the classifier. We have used 5 as our n_neighbors value because we wanted to compare this new result to the original from project 2.

### Part 2.2: Keras Neural Networks

```python
from keras import models
from keras import layers

model = models.Sequential([
  layers.Dense(128, activation='relu', input_shape=(14,)),
  layers.Dropout(0.2),
  layers.Dense(2, activation='softmax')
])
```

```python

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split

# we split the data in training and test.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python

history = model.fit(x_train, y_train, epochs=100)
```

```python
model.evaluate(x_test, y_test)
```

The model we have constructed 3 layers. The first layers i the input layers. This layers consists of 14 input values.
The second layers consists of 128 neurons, which uses the "relu" activation function. This layer also uses a drop out value of 20%. This is bececause we want to make sure that even if there is some missing data, it can still accurately predict the right outcome. The output layers consist of 2 neurons. This is because the outcome is binary(male or female). The output layers uses the softmax function.

## Part 3: Model Training and Validation, Hyperparameter Grid Search

### part 3.1

### Unbalanced 

```python
males = voice_csv[voice_csv['label']=='male']
females = voice_csv[voice_csv['label']=='female']
sixthy_percent_males = males[int(len(males)*0.6):-1]

df = pd.concat([females,sixthy_percent_males])
x = df.iloc[:,0:-1]
y = LabelEncoder().fit_transform(df.iloc[:,-1])

scores = cross_val_score(pipeline1,x, y, cv=10)
sum(scores/len(scores))
print((0.9557191111530337-0.9527087811479978)/0.9527087811479978*100,'% better than the orignal')
```

### rebalanced

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=2)
x, y = sm.fit_sample(x, y)
scores = cross_val_score(pipeline1,x, y, cv=10)
sum(scores/len(scores))

print((0.9766439773903351-0.9527087811479978)/0.9527087811479978*100,'% better than the orignal')
```

The unbalanced dataset performed only marginally better than the orignal. But the rebalanced performed 2.5% better than the original. This might be due to the fact that some of the data is reused.

```python
def plot_confusion_matrix(y_true, y_pred, classes,
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues):
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    # Only use the labels that appear in the data
   
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```
